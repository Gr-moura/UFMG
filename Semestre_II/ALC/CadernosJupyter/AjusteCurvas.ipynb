```python
import matplotlib.pyplot as plt  
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np
from sklearn.datasets import load_diabetes
from numpy import linalg as LA
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

Regressão Linear Simples - Dados Diabetes

```python
sns.set()
X, y = load_diabetes(return_X_y=True, as_frame=True)
X = X[['bmi']]
sns.regplot(x=X, y=y)
plt.title('Regressão Linear Simples - Diabetes', fontsize=20)
plt.xlabel('Índice de Massa Corporal')
plt.ylabel('Métrica de Progressão da Doença')
plt.show()
```

```python
sns.set() 
 
X, y = load_diabetes(return_X_y=True, as_frame=True)
X = X[['bmi']]

lin_reg = LinearRegression().fit(X, y)
sns.regplot(x=X, y=y)
plt.title('Regressão Linear Simples - Diabetes', fontsize=20)
plt.xlabel('Índice de Massa Corporal')
plt.ylabel('Métrica de Progressão da Doença')

plt.show()

print('Intercepto:', lin_reg.intercept_)
print('Coeficientes:', lin_reg.coef_)
print('Coeficiente de Determinação (R2 Score):', lin_reg.score(X, y))
```

Regressão Linear Simples - Túnel de Vento

```python
x = np.array([10, 20, 30, 40, 50, 60, 70, 80]).reshape((-1, 1))
y = np.array([25, 70, 380, 550, 610, 1220, 830, 1450])

plt.scatter(x,y)
plt.title('Experimento: Túnel de Vento')
plt.xlabel('Velocidade (m/s)')
plt.ylabel('Força (N)')


lin_reg = LinearRegression().fit(x, y)
sns.regplot(x=x, y=y,ci=None)
plt.title('Experimento: Túnel de Vento')
plt.xlabel('(Velocidade (m/s)')
plt.ylabel('Força (N)')
#plt.show()
print('Intercepto:', lin_reg.intercept_)
print(' oeficientes:', lin_reg.coef_)
print('Coeficiente de Determinação (R2 Score):', lin_reg.score(x, y))
```

Linearização - Equação de Potência Simples - Túnel de Vento

```python
x = np.array([10, 20, 30, 40, 50, 60, 70, 80]).reshape((-1, 1))
y = np.array([25, 70, 380, 550, 610, 1220, 830, 1450])
xlog = np.log10(x);
ylog = np.log10(y);

plt.scatter(xlog,ylog)
plt.title('Experimento: Túnel de Vento')
plt.xlabel('Velocidade (m/s)')
plt.ylabel('Força (N)')
plt.show()

lin_reg = LinearRegression().fit(xlog, ylog)
sns.regplot(x=xlog, y=ylog,ci=None)
plt.title('Experimento: Túnel de Vento - Linearização')
plt.xlabel('Log(Velocidade (m/s))')
plt.ylabel('Log(Força (N))') 
plt.show()
print('Intercepto:', lin_reg.intercept_)
print('Coeficientes:', lin_reg.coef_)
print('Coeficiente de Determinação (R2 Score):', lin_reg.score(xlog, ylog))

alpha = 10**lin_reg.intercept_
print(alpha)
f = alpha*x**lin_reg.coef_[0]
plt.plot(x,f)
plt.scatter(x,y)
plt.title('Experimento: Túnel de Vento - Equação de Potência')
plt.xlabel('Velocidade (m/s)')
plt.ylabel('Força (N)') 
plt.show()
```

Regressão Linear Simples - População Brasileira


```python
#População Brasileira - Dados transformados
x = np.array([-40, -30, -20, -10, 0, 11, 15, 20, 30])
y = np.array([12.880182, 18.782891, 31.303034, 52.084984,80.436409, 110.990990, 123.076831, 137.953939, 160.925792])
mymodel = np.poly1d(np.polyfit(x, y, 7))
myline = np.linspace(-40, 30, 100)

plt.scatter(x, y)
xano = np.array([1940, 1950, 1960, 1970, 1980, 1991, 1996, 2000, 2010])
plt.plot(myline, mymodel(myline))
plt.xlabel('Ano')
plt.ylabel('População (10^6)')
plt.xticks(x,xano)
plt.show()

R2 = r2_score(y, mymodel(x))
print('Coeficiente de Determinação (R2 Score):',R2)

```

Malcondicionamento das equações normais 

```python
x = np.array([1, 20, 30, 50])
#grau = N-1
N = 4

A = np.vander(x, N)
condA = LA.cond(A)

AT = np.transpose(A)
ATA= np.matmul(AT,A) 
condATA = LA.cond(ATA)


print('Matriz A:', A)
print('Matriz ATA:',ATA)
print('Número de Condição da matriz A', condA)
print('Número  de Condição da matriz ATA', condATA)
 


```
